```python
import os
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:.2f}".format)
```

```python
PROJECT_ROOT = r"C:\Users\CORE\Desktop\customer_segmentation"
DATA_DIR = os.path.join(PROJECT_ROOT, "data")

for file in os.listdir(DATA_DIR):
    if file.lower().endswith(".csv"):
        df = pd.read_csv(os.path.join(DATA_DIR, file), encoding="ISO-8859-1")
        break
    elif file.lower().endswith((".xls", ".xlsx")):
        df = pd.read_excel(os.path.join(DATA_DIR, file))
        break
```

```python
df.head()
```

```python
#why data cleaning is critical - missing values , duplicate records , canceled transactions , returns (negative quantity) , wrong datatypes 


df.isnull().sum()  #checks missing values - Missing values in CustomerID and Possibly missing Description
```

```python
#Customer segmentation is impossible without CustomerID
df = df.dropna(subset=["CustomerID"])  # will keep only valid customers
```

```python
#remove cancelled invoices 
#Invoice numbers starting with “C” = cancelled orders - eg C536379

#WHY TO REMOVE THEM 
# no revenue and distorts frequency and monetary values 
```

```python
df = df[~df["InvoiceNo"].astype(str).str.startswith("C")]
df
```

```python
df.describe()
```

```python
''' u will notice negative quantity , zero or neative unitprice , A valid purchase must have 
quantity > 0 and unitprice > 0 '''

df = df[(df["Quantity"] > 0) & (df["UnitPrice"] > 0)]
df
```

```python
#convert date column 
''' invoicedate is string , we need datetime for recency analyis '''

df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df
```

```python
#remove duplicates 

df = df.drop_duplicates()
df     # Duplicate rows inflate frequency & revenue , Real datasets often contain duplicates
```

```python
#verify clean data 
df.info()
df.describe()
df.isnull().sum()



'''
No missing CustomerID
No negative quantity
No zero price
Correct date format


'''
```

```python
# we converted raw transactional data into business valid , consistent , reliable , ready for analysis data 
```
